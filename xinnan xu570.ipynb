## Methods
For predicting continous outcomes, for this case the house price, we will apply a machine learning algorithm called Linear regression. Linear regression is a machine learning algorithm that is used for prediction and forecasting, in addition, linear regression shows the impact an independent variable has on dependent variable. Independent variable also known as predictor variables are the variables that explain or predict the outcome of a dependent variable, on the other hand, a dependent variable is the variable we want to predict.

Linear regression uses the least square method.

For, this study, price is the dependent variable and the independent variables are the number of bedrooms, number of bathrooms, size of living room in square feets, the size of lot in square feet, and the house condition.

We'll carry out a step model, that is, we'll analyze the data using different treatments/features and find out the one that is a perfect fit in predicting the outcome. However, before performing the actual analysis we can carry out correlation analysis to find out the association between these variables.

```python
Corr_Matrix=round(Housing.corr(),2)  # Finds the correlation and rounding it to 2 decimal places
print(Corr_Matrix)
```

```python
# We'll have to split the dataset into a training and test set first, then we'll build the regression model and make predictions.
from sklearn.model_selection import train_test_split
y= Housing.price
X= Housing.drop('price', axis=1)

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

# Now we'll fit the regression model; this first model has all the features
from sklearn.linear_model import LinearRegression
regr= LinearRegression()
results=regr.fit(X_train,y_train)
attributes_coefficients = pd.DataFrame(regr.coef_, X.columns, columns=['Coefficient'])  
attributes_coefficients
```

Using the coefficients we can make inferences regarding the impact of the predictor variables on the dependent variable. Here it can be seen that:
* As the number of bedrooms decrease by one unit the price increases by 65516 dollars.
* As the number of bathrooms increase by one unit the price of the house increases as well, by 22573 dollars.
* As the size of the living room increases by one square foot, the house price increases by 282 dollars
* As the size of the lot decreases by one square foot, the price increases by 0.9 dollars
* As the house condition improves with unit, the price increases by 58323 dollars.

```python
# We will now make the predictions; predictions are made on test set, thereafter, we'll compare them with the actual prices
y_pred = regr.predict(X_test)
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
comparison.head(n=10)
```

```python
# Let's asses whether our model is a perfect fit or not, we'll use the coefficient of determination to do this;
from sklearn import metrics 
print('Adjusted R-square:', metrics.r2_score(y_test, y_pred))
```

```python
# Fit a new model with only bathrooms, sqft_living, and condition as the only predictor variables
y= Housing.price
X= Housing.drop(labels=['price', 'sqft_lot'], axis=1)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
regr2= LinearRegression()
results=regr2.fit(X_train,y_train)
attributes_coefficients = pd.DataFrame(regr2.coef_, X.columns, columns=['Coefficient'])  
attributes_coefficients
```

```python
# Next we make the predictions
y_pred = regr2.predict(X_test)
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
comparison.head(n=10)
```

```python
# Now we evaluate model2 performance
print('Adjusted R-square:', metrics.r2_score(y_test, y_pred))
```

## Findings
From the analysis it can be evaluated that the listed features, i.e, the number of bedrooms the number of bathrooms, the size of the living room, the size of the lot, and the house condition all have some impact on the price of the house. However, these impacts vary, some have positive impact, that is as their value increases the price of the house also tend to increase as in the case of the number of bathrooms, size of the living room, and the condition of the house, also, some features have an inverse relationship with the house as in the case of the number of bedrooms and the lot size.

In addition, it can be noticed that model 1 and 2 all yield differnt results. In the first model we included all the features and in the second mode we dropped some (sqft_lot, was dropped in model 2). To asses which model is a perfect fit for making the predictions, we determined their R-square values, the first model had a value of 50.3% while the second one had a value 0f 50.7%. Therefore, for making predictions we would prefer the second odel to the first model.
